```python
import numpy as np
import matplotlib.pyplot as plt
# use matplotlib inline
plt.ion()
```

```python
# x_train is the input variable
# y_train is the output variable
X_train = np.array([1, 2])
y_train = np.array([300, 500])
print(f"X Training Data: {X_train}")
print(f"y Training Data: {y_train}")
```

```python
# m is the number of training examples
m = X_train.shape[0]
print("Number of training examples:", m)
```

```python
# Plot the data points
plt.scatter(X_train, y_train, marker="x", c='r')
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel("Price (in 1000s of dollars)")
# Set the x-axis label
plt.xlabel("Size (1000 sqft)")
plt.show()
```

```python
# Parameters
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model(x, w, b):
    """Computes the prediction of a linear model
    
    Args:
    X -- data
    w, b params
    Return: y target values
    """
    m = x.shape[0]
    f_wb = np.zeros(m)

    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
# Prediction

y = compute_model(X_train, w, b)
print(f"Predicted value is: {y}")

# Plot our model prediction
plt.plot(X_train, y, c="b", label="Our Prediction")
plt.scatter(X_train, y_train, marker="x", c="r", label="Actual Values")
plt.title("Housing Prices")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (1000 sqft)")
plt.legend()
plt.show()
```
